```python
import pandas as pd
```

```python
path = r'/Users/nikolaibeckjensen/Library/CloudStorage/OneDrive-DanmarksTekniskeUniversitet/Virk2Vec/data/processed/corpus/global_set/sentences/train/part.0.parquet'
```

```python
df = pd.read_parquet(path)
```

```python
df
```

```python
df['token_count'] = df['SENTENCE'].apply(lambda x: len(x.split()))
```

```python
grouped = df.groupby("CVR").agg({'token_count':'sum', 'AGE':'count'}).rename(columns={'token_count':'token_count_sum', 'AGE':'sentence_count'})
grouped['avg_token_count'] = grouped['token_count_sum'] / grouped['sentence_count']
```
